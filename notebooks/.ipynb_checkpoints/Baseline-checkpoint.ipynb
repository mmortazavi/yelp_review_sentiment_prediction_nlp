```python
import os
import sys
import pandas as pd 
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier

module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)

from yelp import baseline
from yelp.preprocessing import fast_text_clean
from yelp.feature_generation import text_to_features
from yelp.feature_generation import top_feats_by_class, plot_classfeats_h

%matplotlib inline
```

```python
classInit = baseline.BaseEstimator(path = '/Users/majidmortazavi/data_sciences/case_studies/NewYorker/data', filename = 'processed_data_0.1.parquet.gzip',method = 'countVectorizer', classifier = MultinomialNB)
```

```python
data = classInit.data

data
```

```python
data = classInit.data

df_feat_wordcount, _ = text_to_features(data['text'], verbose=0, analyzer=fast_text_clean, features='countVectorizer')

features = df_feat_wordcount.columns.tolist()
```

```python
dfs_wordcount = top_feats_by_class(df_feat_wordcount.values, data['stars'], features, min_value=0.01, top_n=15)
plot_classfeats_h(dfs_wordcount, feature_name='Average Count')
```

```python
classInit.run()
```
